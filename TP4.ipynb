### Question 1

Définir une fonction tree_summary qui prend un argument un arbre de décision entraîné et qui affiche son score d’apprentissage, son score de test, sa profondeur et son nombre de nœuds. Appliquer cette fonction à l’arbre dt construit ci-dessus. Commenter les résultats.

```python
def tree_summary(arbre):
    
    score_apprentissage = arbre.score(X_train, y_train)
    score_test = arbre.score(X_test, y_test)
    profondeur =  arbre.get_depth()
    nb_noeuds = arbre.tree_.node_count

    print("\n Score d'apprentissage :", score_apprentissage, "\n Score de test", score_test, "\n Profondeur :", profondeur, "\n Nombre de noeuds :", nb_noeuds, "\n")
    
    return score_apprentissage, score_test, profondeur, nb_noeuds
```

```python
tree_summary(dt)
```

Beaucoup de noeuds. Grande profondeur. Score de test moins élevé que score d'entrainement (normal par construction. Score d'apprentissage = 1 et score de test pas aussi élevé => Surapprentissage

### Question 4

A l’aide d’une 10-validation croisée, choisir le meilleur paramètre pour la profondeur maximale de l’arbre de décision.

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor

# Définir les valeurs à tester pour le paramètre de profondeur maximale
param_grid = {'max_depth': [3, 5, 7, 9, 11]}

# Initialiser le modèle d'arbre de décision
dt = DecisionTreeRegressor(random_state=0)

# Recherche du meilleur paramètre de profondeur maximale avec validation croisée
grid_search = GridSearchCV(dt, param_grid, cv=10, scoring='neg_mean_squared_error')
grid_search.fit(X_train, y_train)

# Obtention des meilleurs paramètres et du meilleur score
best_params = grid_search.best_params_
best_score = -grid_search.best_score_

print("Meilleur paramètre de profondeur maximale:", best_params)
print("Meilleur score (RMSE) avec validation croisée:", best_score)
```

### Question 5

Définir une fonction r2_score qui prend un argument deux arrays y_true et y_predict contenant respectivement, pour un échantillon donné, les vraies étiquettes et les étiquettes prédites par un prédicteur, et qui revoie le score R2 correspondant.

```python
def r2_score(y_true, y_predict):

    r2 = 1 - (np.sum((y_true - y_predict)**2)/(np.sum((y_true - np.mean(y_true))**2)))
    
    return r2
```
